```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.zeros(len(x))
for z in x :
     y = np.sin(x)

## filter
a = np.linspace(-1,1,3)
f = [1 , 0 , -1]
print(a , f , x , y )
```

```python
plt.plot(x,y, label="sin wave")
plt.plot(a,f , label = "filter")
```

```python
#now multiply to start the convolution
c = np.zeros(len(y))

for i in range(len(y)):
    z = np.zeros(len(f))
    for j in range(len(z)) :
        if i+j < len(y) :
            z[j] = y[i+j] * f[j]
    c[i] = np.sum(z)

print(c)
```

```python
plt.plot(x,c,label="convultion")
plt.plot(x,y)
plt.grid()
```
